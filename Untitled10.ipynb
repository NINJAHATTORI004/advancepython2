```python
#encapsulation
class super:
  def __init__(self):
    self.__value1=20
    self.__value2=40
  def display(self):
    print(self.__value1)
    print(self.__value2)
class sub():
  def show(self):
    print(self.__value1)
    print(self.__value2)
ob=sub()
ob.show()
```

```python
class super():
  def __init__(self):
    self.value1=20
    self.value2=40
  def display(self):
    print(self.value1)
    print(self.value2)
ob1=super()
ob1.display()
```
