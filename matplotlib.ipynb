```python
#data visulization
import matplotlib.pyplot as plt
%matplotlib inline
x=[11,12,13,14,15,16,17]
y=[3423,44,330000000000000000000,76,56,-56,55]
plt.plot(x,y)
```

```python
#graph using arrays


import numpy as np

a1=np.array([[11,12],[13,14],[15,16]])

a2=np.array([[-1,2],[3,-4],[5,-6]])
plt.plot(a1,a2,x,y,color="green",linewidth="1",linestyle="dotted")
```

```python
#line width
plt.xlabel("day")
plt.ylabel("temperature")
plt.title("weather")
plt.plot(x,y,color="red")
```

```python
plt.plot(y,x,color="black",linestyle="dotted", linewidth="5")
```

```python
plt.plot(a1,a2)
```

```python
plt.plot(x,y,a1,marker="+",color="purple",linestyle="dashdot",alpha=1,label="min")
```
