```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt

# Load the dataset
california_housing = fetch_california_housing()

# Convert to DataFrame for easier analysis
df = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)

# Add target column (median house value)
df['MedHouseVal'] = california_housing.target

# 1. Total number of districts
total_districts = df.shape[0]
print("Total number of districts:", total_districts)

# 2. Average median house value across all districts
average_median_house_value = df['MedHouseVal'].mean()
print("Average Median House Value:", average_median_house_value)

# 3. District with the highest median income
highest_median_income_district = df.loc[df['MedInc'].idxmax()]
print("District with the highest median income:\n", highest_median_income_district)

# 4. Average house age in districts with median house value above $300,000
filtered_df = df[df['MedHouseVal'] > 300000]
average_house_age = filtered_df['HouseAge'].mean()
print("Average House Age in districts with house value above $300,000:", average_house_age)

# 5. Create a histogram of median house values
plt.hist(df['MedHouseVal'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Median House Values')
plt.xlabel('Median House Value ($100,000s)')
plt.ylabel('Number of Districts')
plt.show()
```
