# Axes Labels, Legend, Grid

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t)
```

```python
#MAX,MIN and Average Teemperature all in one chart
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
# In this chart we do't know what this x-axis and y-axis
```

```python
# set axes labels and chart title
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')


plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
# Show legend, In order to get legend we first need to set labels
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='best')
```

```python
# Legend at different location with shadow enabled and fontsize set to large
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='upper right',shadow=True,fontsize='large')
```

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='upper right', fontsize="large",shadow=True)#fontsize="small"
```

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='upper right', fontsize="large",shadow=True)#fontsize="small"
plt.grid()
```

# HISTOGRAMS: BAR CHARTS HAVING FREQUENCY OF A VARIABLE

TO ANSWER SUCH QUESTIONS FOLLOWING CHART WILL BE USEFUL

# 

```python
import matplotlib.pyplot as plt
%matplotlib inline
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
# When we want to plt Histogram we just need single dimension array to be plotted on x axis: Y AXIS CONTAINS FREQUENCY
plt.hist(blood_sugar) # by default number of bins is set to 10
```

```python
plt.hist(blood_sugar,rwidth=0.5)
```

```python
plt.hist(blood_sugar,rwidth=0.5,bins=4)
```

# Histogram showing normal, pre-diabetic and diabetic patients distribution

80-100: Normal

100-125: Pre-diabetic

80-100: Diabetic

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,190], rwidth=0.95, color='g')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```

# PIE CHARTS

# TRACKING HOW MUCH PERCENTAGE OR PI INDIVISUAL ITEM IS TAKING OUT OF TOTAL EXPENSE, HAVING A PIE CHART HELPS A LOT

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

# Draw pie chart as perfect circle

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=.5)
plt.show()
```

# Saving matpotlib chart to a file

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=.5)
plt.savefig("piechart.jpg", bbox_inches="tight", pad_inches=1, transparent=True)
```
