```python
class Circle():
  def __init__(self,r):
    self.radius=r
  def area(self):
    return self.radius**2*3.14
  def perimeter(self):
    return 2*self.radius*3.14
NewCircle=Circle(8)
print("new circle area:",NewCircle.area())
print("new circle perimeter:",NewCircle.perimeter())
```

```python
#practical program 3
class Rectangle():
  def __init__(self,l,b):
    self.length=l
    self.breath=b
  def area(self):
    return self.length*self.breath
  def perimeter(self):
    return 2*(self.length+self.breath)
NewRectangle=Rectangle(10,300)
print("new rectangle area:",NewRectangle.area())
print("new rectangle perimeter:",NewRectangle.perimeter())
```

```python
import random

# Questions database
questions = {
    "Which is the capital of India?": ["Mumbai", "Delhi", "Kolkata", "Chennai", 2],
    "What is the largest planet in our solar system?": ["Jupiter", "Saturn", "Neptune", "Uranus", 1],
    "Who painted the Mona Lisa?": ["Leonardo da Vinci", "Vincent van Gogh", "Pablo Picasso", "Michelangelo", 1],
    # Add more questions here
}

# Game function
def play_kbc():
    score = 0
    
    # Shuffle the questions
    question_list = list(questions.items())
    random.shuffle(question_list)
    
    for question, options in question_list:
        print(question)
        for i, option in enumerate(options[:-1]):
            print(f"{i+1}. {option}")
        
        answer = int(input("Enter your answer (1-4): "))
        
        # Check the answer
        if answer == options[-1]:
            print("Correct answer!")
            score += 1
        else:
            print("Wrong answer!")
            break
    
    print(f"\nGame over! Your final score is {score} out of {len(questions)}.")

# Run the game
play_kbc()
```

```python
import matplotlib.pyplot as plt

def print_bar_graph(x_labels, data):
    # Set up the figure and axis
    fig, ax = plt.subplots()
    
    # Plot the data as a bar graph
    ax.bar(x_labels, data)
    
    # Add labels and title
    ax.set_xlabel("Categories")
    ax.set_ylabel("Values")
    ax.set_title("Bar Graph")
    
    # Display the bar graph
    plt.show()

# Example usage
x_labels = ["Category A", "Category AA", "Category B", "Category BB"]  # Labels for x-axis
data = [10, 24, 16, 30]  # Corresponding data for each category
print_bar_graph(x_labels, data)
```
