```python
import pandas as pd
import numpy as np
import seaborn as sns

# Load tips dataset
tips = sns.load_dataset('tips')

# 1. What is the average billing of Restaurant?
avg_billing = tips['total_bill'].mean()
print(f"\n1. Average billing of Restaurant: {avg_billing:.2f}")

# 2. What is the median price of total_bill?
median_bill = tips['total_bill'].median()
print(f"\n2. Median price of total_bill: {median_bill:.2f}")

# 3. Which size of tip is mostly given to waiters?
most_common_tip = tips['tip'].mode()[0]
print(f"\n3. Most common tip given to waiters: {most_common_tip:.2f}")

# 4. Find the variance of tips.
variance_tips = tips['tip'].var()
print(f"\n4. Variance of tips: {variance_tips:.2f}")

# 5. Find Standard Deviation of total_bill.
std_total_bill = tips['total_bill'].std()
print(f"\n5. Standard Deviation of total_bill: {std_total_bill:.2f}")
```
